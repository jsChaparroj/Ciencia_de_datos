Algunas variables presentan valores ausentes, las cuales están reportadas como: No reportado, no reporta, "-" entre otros.
Este problema se soluciona cambiando estos valores como nan


```python
def limpiar_paraNa(x):
  if type(x) is str:
    if "NO " in x:
      x = np.nan
      return x
    else:
     if "-" in x:
       x = np.nan
       return x
  return x
```

```python
datos  = datos.applymap(limpiar_paraNa)
```
