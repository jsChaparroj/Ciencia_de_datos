### 2. Obtener una base de datos por municipios, que contenga la cantidad de delitos por tem´atica, porcentaje de mujeres victimas y porcentaje de delitos en zonas rural

```python
print(datos["SEXO"].unique())
print(datos["ZONA"].unique())
```

```python
def limpiar_sexo(x):
    if x is np.nan:
        return(x)
    if "NO REPOR" in x:
        x = np.nan
        return(x)
    if "-" in x:
        x = np.nan
        return(x)
    else:
        return(x)
    
datos["SEXO"] = datos["SEXO"].apply(limpiar_sexo)
```

```python
filtro1 = datos[(datos["SEXO"] == "FEMENINO") & (datos["ZONA"] == "RURAL")]
filtro1.sample(5)
```

```python
df1 = pd.pivot_table(filtro1,index = ["MUNICIPIO","TEMÁTICA"] ,values=["SEXO","ZONA"]  ,  aggfunc="count")
df1
```
